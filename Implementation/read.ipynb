```python
import igraph 
import csv
import matplotlib
import unicodedata
```

```python
def readVertices(path, graph, addNames):
    names = []
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count = 0
        for row in csv_reader:
            if line_count>0:
                graph.add_vertices(row[2])
                string=row[0]
                string=unicodedata.normalize('NFKD', string).replace(u'Å‚', 'l').encode('ascii', 'ignore')
                names.append(string.decode('utf-8'))
            line_count+=1
    if addNames:
        graph.vs["name"]=names
    
```

```python
def readEdges(path, graph, addEdges):
    weights=[]
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count2 = 0
        for row in csv_reader:
            if line_count2>0:
                try:
                    graph.add_edges([( int(row[0]), int(row[1]))])
                    weights.append(int(row[2]))
                except:
                    {}
            line_count2+=1
    if addEdges:
        graph.es['weight'] =weights
```

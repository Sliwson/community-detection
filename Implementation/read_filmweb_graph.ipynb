```python
import igraph 
import csv
import unicodedata
```

```python
def readVertices(path, graph):
    names = []
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count = 0
        for row in csv_reader:
            if line_count>0:
                graph.add_vertex(name=row[2])
            line_count+=1

    
```

```python
def readEdges(path, graph, addCommonMoviesAsWeight, addYearsKnownAsWeight):
    commonMovies=[]
    years=[]
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count2 = 0
        for row in csv_reader:
            if line_count2>0:
                try:
                    label1=graph.vs["name"].index(row[0])
                    label2=graph.vs["name"].index(row[1])
                    graph.add_edges([(label1,label2)])
                    years.append(2020-int(row[2]))
                    commonMovies.append(int(row[3]))
                except:
                    {
                        print(row[0],row[1],"===============")
                    }
            line_count2+=1

    if addCommonMoviesAsWeight:
        graph.es['weight'] =commonMovies
    if addYearsKnownAsWeight:
        graph.es['weight']=years
```

```python
def getDates(path):
    dates=[]
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count2 = 0
        for row in csv_reader:
            if line_count2>0:
                dates.append(int(row[2]))
            line_count2+=1
    return dates
```

```python
def getCommonMoviesCount(path):
    commonMoviesCount=[]
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count2 = 0
        for row in csv_reader:
            if line_count2>0:
              commonMoviesCount.append(int(row[3]))
            line_count2+=1
    return commonMoviesCount
```

```python
def getNames(path):
    names = []
    with open(path, mode='r', encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        line_count = 0
        for row in csv_reader:
            if line_count>0:
                string=row[0]
                string=unicodedata.normalize('NFKD', string).replace(u'Å‚', 'l').encode('ascii', 'ignore')
                names.append(string.decode('utf-8'))
            line_count+=1
    return names
    
```
